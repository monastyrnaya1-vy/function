```python
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
```

```python
directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}
    
```

```python
def p(num): 
    for i in documents:
        if num == i.get('number'):
            return i.get('name')
        
    return "Документа не существует"
    

def s(doc):
    for i,v in directories.items():
        if doc in v:
            return i
        elif doc == '':
            return 3
            
    return "Документа не существует"
     
def list_():
    for i in documents:
        print(f'{i.get("type")} "{i.get("number")}" "{i.get("name")}"')
              

   
def ass(shelf):
    if shelf in directories.keys():
        return 'такая полка уже существует'
    else:
        directories.setdefault(shelf,[])
        return directories.copy()             

       
def de(delete):
    for i in documents:
        if delete in i.values():
            i.clear()
            print(documents)
            for i, v in directories.items():
                for el in v:
                    el.split(",")
                    if delete == el:
                        v.remove(delete)
                        return v
    return 'Неверный номер документа'
                    
    
              
def docs_shelvs(docs, shelvs):
    for k, v in directories.items():
        if docs in v and shelvs in directories.keys():
            directories[k].remove(docs)
            directories[shelvs].append(docs)
            return directories   
    return 'Неверные данные'     
       
              
def add(doc, doc_t, fio, shelf):
    if shelf in directories:
        documents.append({'type': doc_t, 'number': doc, 'name': fio})
        directories[shelf].append(doc)
        return documents, directories
    else:
        return "Полки не существует"
            
    
    


            
            
            


    
```

```python
comand = input('Введите команду ')

def main(comand):
    
    while True:
        if comand == 'p':
            num = input('Введите номер документа ')
            return p(num)
        elif comand == 's':
            doc = input('Введите номер документа ')
            return s(doc)
        elif comand == 'l': 
            return list_()
        elif comand == 'as':
            shelf = input('Введите номер новой полки: ')
            return ass(shelf)
        elif comand == 'd':
            delete = input('Введите номер документа: ')
            return de(delete)
        elif comand == 'm':
            docs = input('Введите номер документа: ')
            shelvs = input('Введите номер полки: ')
            return docs_shelvs(docs, shelvs)
        
        elif comand == 'a':
            doc = input('Введите номер документа: ')
            doc_t = input('Введите тип документа: ')
            fio = input('Введите ФИО владельца: ')
            shelf = input('Введите номер полки: ')
            return add(doc, doc_t, fio, shelf)
        elif comand == 'exit':
            break
    
main(comand)        
    
        
```
